```python
 #Two arrays of data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 9, 16, 25]

# Creating subplots
plt.subplot(1, 2, 1)  # 1 row, 2 columns, subplot 1
plt.plot(x, y1, 'r--')  # 'r--' specifies red dashed line
plt.title('Plot 1')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, subplot 2
plt.plot(x, y2, 'g-')  # 'g-' specifies green solid line
plt.title('Plot 2')

plt.tight_layout()  # Adjust layout to prevent overlap of subplots
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.array([1, 2, 6, 8])
y1 = np.array([3, 8, 2, 10])
y2 = np.array([1, 4, 9, 16])

# Creating subplots
plt.subplot(1, 2, 1)  # 1 row, 2 columns, subplot 1
plt.plot(x, y1, 'r--')  # 'r--' specifies red dashed line
plt.title('Dotted Line plot')

plt.subplot(1, 2, 2)  # 1 row, 2 columns, subplot 2
plt.scatter(y1, y2, color='g')  # it plots the values in scatter dots
plt.title('Scatter Plot')

plt.tight_layout()  # Adjust layout to prevent overlap of subplots
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

# Read the CSV file into a Pandas DataFrame
a = pd.read_csv('pandasss.csv')

# Extract values of the 2nd and 3rd columns
column_2_values = a.iloc[:, 2]  # 2nd column
column_3_values = a.iloc[:, 3]  # 3rd column



# Plotting the bar graph
plt.figure(figsize=(25, 15))  # Adjust width and height as needed fro the size of graph
plt.scatter(column_2_values, column_3_values,color='r', linewidth=10)

# Adding labels and title
plt.xlabel('Student name')
plt.ylabel('Student age')
plt.title('Bar Graph Student name and age')
plt.grid('true')

# Display the plot
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# Read the CSV file into a Pandas DataFrame
a = pd.read_csv('pandasss.csv')

# Extract values of the "student name" and "student age" columns
student_names = a["Student Name"]
student_ages = a["Student Age"]

# Create an array for the x-axis (index values)
x = range(len(student_names))

# Increase the width of the bars
bar_width = 0.5  # Adjust this value as needed

# Set figure size to extend the width
plt.figure(figsize=(10, 6))  # Adjust width and height as needed

plt.bar(x, student_ages, width=bar_width)

# Adding labels and title
plt.xlabel('Student Name')
plt.ylabel('Student Age')
plt.title('Bar Graph of Student Ages')

# Rotating x-axis labels for better readability
plt.xticks(x, student_names, rotation='vertical')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

x = ["python", "java", "react", "c++"]
y = [20, 30, 40, 50]
colors = ['r', 'g', 'b', 'orange']  # List of colors for each bar

plt.bar(x, y, color=colors, label="popularity",edgecolor="b")
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title('Bar Graph Popularity of Programming Languages')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
A = ["Python", "JavaScript", "C++", ".Net"]
B = [20, 30, 40, 50]
C = [60, 70, 80, 90]

# Define the width of the bars
width = 0.3

# Define the x locations for the groups
x = range(len(A))
plt.figure(figsize=(10, 6))
# Plot bars for B
plt.bar(x, B, width=width, label='shareholders', color='blue')

# Plot bars for C, shifting x positions to right
plt.bar([i + width for i in x], C, width=width, label='popularity', color='orange')

# Customize the plot
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Languages')
plt.xticks([i + width / 2 for i in x], A)  # Adjust x-axis labels
plt.legend()

# Show plot
plt.show()
```

```python
#stem graph 
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[0,3,4,5,6]
plt.stem(x,y,linefmt=":",markerfmt="r*",bottom=0,basefmt="g",label="python",orientation="horizontal")
plt.legend() #graph bhitra label dekhauna lai 
plt.show()
```

```python
#piechart
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=["c","c++","Python","Javascript"]
ex=[0.4,0.3,0.0,0.2]
c=["r","g","b","orange"]
plt.pie(x,labels=y, explode=ex, colors=c, autopct="%0.1f%%",shadow=True) #explode is used to provide the gap of the pie and autocpt is used to give the decimal value after point.
plt.show()
```

```python
import matplotlib.pyplot as plt 
x=[1,2,3,4,5]
area1=[2,3,4,5,6]
area2=[1,2,3,4,5]
area3=[6,5,4,3,2]
l=["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels=l,colors=["r","g","m"])
plt.title("Stack Plot")
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[5,4,3,2,1]
plt.step(x,y, marker='o', label="Step plot")
plt.title("Step plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid()
plt.show()
```

```python
#Fill between plot
x=[1,2,3,4,5]
y=[5,4,3,2,1]
plt.plot(x,y, color='r', label="Fill between Plot") 
plt.fill_between(x=[4,2],y1=4,y2=2)
plt.title("FIll between Plot")
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```
