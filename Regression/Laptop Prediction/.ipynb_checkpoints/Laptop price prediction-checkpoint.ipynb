```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
corr_matrix = df.corr()
```

```python
# Plot the correlation matrix using seaborn with enhanced visualization
plt.figure(figsize=(20,15))  # Increase figure size for better readability
sns.set(font_scale=1.2)  # Increase the font scale for better readability
heatmap = sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap='YlGnBu',  # Changed the color map to 'YlGnBu' for a different look
    fmt='.2f', 
    vmin=-1, 
    vmax=1, 
    linewidths=.5,  # Add grid lines for better separation
    cbar_kws={"shrink": .8}  # Shrink the color bar for better aesthetics
)
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=45, horizontalalignment='right')
heatmap.set_yticklabels(heatmap.get_yticklabels(), rotation=0)
plt.title('Correlation Matrix', fontsize=20)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="Company", y="Price", data=dfOriginal)
plt.xlabel('Company')
plt.ylabel('Laptop Price')
plt.title('laptop price by company')
plt.xticks(rotation=45)
plt.show()            
```

```python
# Violin plot of laptop prices by company
plt.figure(figsize=(12, 8))
sns.violinplot(x='Company', y='Price', data=dfOriginal)
plt.title('Laptop Price by Company')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot of average laptop price by company
plt.figure(figsize=(12, 8))
avg_price_by_company = dfOriginal.groupby('Company')['Price'].mean().reset_index()
sns.barplot(x='Company', y='Price', data=avg_price_by_company)
plt.title('Average Laptop Price by Company')
plt.xticks(rotation=45)
plt.show()
```

```python
# Swarm plot of laptop prices by company
plt.figure(figsize=(12, 8))
sns.swarmplot(x='Company', y='Price', data=dfOriginal)
plt.title('Laptop Price by Company')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average price by company
avg_price_by_company = dfOriginal.groupby('Company')['Price'].mean().reset_index()

# Create subplots: 2 rows, 3 columns
fig, axes = plt.subplots(2, 3, figsize=(50, 25))

# Box Plot
sns.boxplot(data=dfOriginal, x='Company', y='Price', ax=axes[0, 0])
axes[0, 0].set_title('Box Plot of Laptop Prices by Company')

# Violin Plot
sns.violinplot(data=dfOriginal, x='Company', y='Price', ax=axes[0, 1])
axes[0, 1].set_title('Violin Plot of Laptop Prices by Company')

# Bar Plot of Average Prices
sns.barplot(data=avg_price_by_company, x='Company', y='Price', ax=axes[0, 2])
axes[0, 2].set_title('Bar Plot of Average Laptop Prices by Company')

# Histogram of Prices
sns.histplot(data=dfOriginal, x='Price', ax=axes[1, 0])
axes[1, 0].set_title('Histogram of Laptop Prices')

# Scatter Plot of Ram vs Price
sns.scatterplot(data=dfOriginal, x='Ram', y='Price', hue='Company', ax=axes[1, 1])
axes[1, 1].set_title('Scatter Plot of Ram vs Price')

# Line Plot of Average Price over Companies
sns.lineplot(data=avg_price_by_company, x='Company', y='Price', ax=axes[1, 2])
axes[1, 2].set_title('Line Plot of Average Laptop Prices by Company')

# Adjust layout
plt.tight_layout()
plt.suptitle("Various Plots of Laptop Prices by Company", y=1.02, fontsize=20)

# Show the figure in Jupyter Notebook
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Compute the correlation matrix
correlation_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(16, 12))

# Create the heatmap with annotations
heatmap = sns.heatmap(
    correlation_matrix, 
    annot=True, 
    fmt=".2f", 
    cmap='viridis', 
    cbar_kws={'shrink': .8}, 
    linewidths=1, 
    linecolor='black', 
    annot_kws={"size": 12}
)

# Adjust the font size for the axes
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Set the title
plt.title('Heatmap of Feature Correlations', fontsize=24, pad=20)

# Show the figure
plt.show()
```
