```python
# Constructor : Constructors in python class method is used for creating and initializing 
#parameterized  Constructor
class Student:
    def __init__(self,name,age,marks,gender,phone):
        self.name=name #__init__ self is used to pass the parameters from one function to another function between the same class 
        self.age=age
        self.marks=marks
        self.gender=gender
        self.phone=phone
    def StudentDetails(self):
        print("name:", self.name)
        print("age:" ,self.age)
        print("marks:",self.marks)
        print("gender:",self.gender)
        print("phone:",self.phone)
        
person=Student("asbin",19,4,"male",9824235695)
person.StudentDetails()
```

```python
#Inheritance 
#Single Inheritance 
class A:
    def A(self):
        print("Inherited Property")
class B(A):
    def B(self):
        print("Sons Property")
prop= B()
prop.A()


#Multi Inheritance
```

```python
#Multi levelInheritance
class A:
    def A(self):
        print("Inherited Property")
class B(A):
    def B(self):
        print("Sons Property")
class C(A):
    def c(self):
        print("Next generation property")
prop=C()
prop.A()
```

```python
#Polymorphism : Same function name(but different signatures) used for overloading and overriding
class Area:
    def cal_area(self,x=None,y=None):
        if x!=None and y!=None:
            print(x*y)
        elif x!=None:
            print(x*x)
        else:
            print("nothing")
obj=Area()
obj.cal_area()
obj.cal_area(10)
obj.cal_area(10,20)

```

```python
class A:
    def show_data(self):
        print("I am in Class A")
class B(A):
    def show_data(self):
        super().show_data()
        print("I am class B")
obj=B()
obj.show_data()
        
```

```python
#Here we have a base class A and B & c are the sub class, inhereting from A class
#D inherits from B  & C also getting the access to methods of class A. So this is an example of a multi-level inheritance 
class A:
    def name(self,name):
        print("The name is", name)
class B(A):
    def midname(self,midname):
        print("The midname is ", midname)
class c(B):
    def lastname(self,lname):
        print("The last name is ", lname)
class D(c):
    def fullname(self,fullname):
        print("The full name is ", fullname)
obj=D()
obj.name("Sandesh")
obj.midname("thapa")
obj.lastname("magar")
```

```python
#Find the prime numbers between 1-50
print("Prime numbers between 1 and 50:")
for num in range(2, 51):
    if all(num % i != 0 for i in range(2, int(num ** 0.5) + 1)):
        print(num)
```
